1) Explain why decision trees are non-parametric models.

    A Non-parametric method means that the model is created on observed data. 
    For example, in order to give someone a loan at a bank, they will have to answer to a list of questions. 
    These questions are sequential in that one correct answer leads to the next. It is the basis of a decision tree.

2.	Create a Python file.  Use this code to read and preprocess the data.  (see number 7 for how to install graphviz)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn import tree
import graphviz

# Metrics
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

df = pd.read_csv('https://raw.githubusercontent.com/grbruns/cst383/master/College.csv', index_col=0)
```

3.	Convert the 'Private' column to an numeric column with values 0 and 1 (1 for private colleges).

```python
df.Private.replace(('Yes', 'No'), (1, 0), inplace=True)
df.head()
```

5.	We will try to predict whether a college is public or private.  Select a few predictors, create NumPy arrays X and y, and then do a training/test split.  Try hard to remember how to do this from memory.  If you can't, refer to the hints.

```python
pred = ['Outstate', 'F.Undergrad']
X = df[pred].values
y = df['Private'].values

X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

6.	Train a tree classifier using Scikit-Learn's DecisionTreeClassifier.  Use the training data you created in the previous step.

```python
# Create a Decision Tree Classifier Object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifier
clf_train = clf.fit(X_train, y_train)

# Predict the response for test dataset 
y_pred = clf_train.predict(X_test)
```

7.	Install graphviz by entering conda install python-graphviz at the Anaconda prompt.  Then plot your tree using graphviz.  Try playing with some of the options of export_graphviz(). 

```python
# Show decision tree as a text
text_representation = tree.export_text(clf_train)
print(text_representation)
```

```python
# Plot a tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf_train, 
                   feature_names=pred,
                   filled=True)
```

```python
# Plot with graphviz
d = tree.export_graphviz(clf_train, out_file=None, feature_names=pred, filled=True)
graph = graphviz.Source(d, format="png")
graph
```

8.	Use your classification tree to predict whether examples in your test data are public or private.  Compute the confusion matrix and the accuracy of your predictions..

```python
# Model Accuracy, how often is the classifier correct? (Compare actual vs predicted)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
